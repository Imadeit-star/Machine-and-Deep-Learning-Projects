### This is a Case Study on Employement Attrition Using Naive Beyesian Model
### This Project makes predictions on the state at which employees leave a company or stay based on voluntary or involuntary reasons

```python
# Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
# Loading the dataset
Employee = pd.read_csv("Employee_Attrition.csv")
```

```python
Employee.head() # Prints first 5 rows of the dataset
```

```python
# Dropping off the employeenumber
Employee = Employee.drop(columns = "employeenumber", axis = 1)
```

```python
# 0----> attrition incdicates that the employee stays with the company.
# 1----> attrition indicates that the employee leaves the company.
Employee.head()
```

```python
# Separating the dataset into features and target
x = Employee.drop(columns = "attrition", axis = 1)
y = Employee['attrition'] 
```

```python
x.head()
```

```python
y = pd.DataFrame(y)
```

```python
y.head()
```

```python
# Converting the boolean values to values between 1 and 0
def scale_true_false_columns(df):

    # Identify True/False columns
    true_false_cols = df.select_dtypes(include=['bool']).columns

    # Create a MinMaxScaler
    scaler = MinMaxScaler()

    # Scale the True/False columns
    df[true_false_cols] = scaler.fit_transform(df[true_false_cols])

    return df  # Return the dataframe with scaled True/False columns
```

```python
y = scale_true_false_columns(y)
```

```python
y.head()
```

```python
# splitting our dataset into train and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.naive_bayes import GaussianNB # Importing the model
```

```python
model = GaussianNB()
```

```python
# Fitting our dataset to the model
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
```

```python
# Checking for accuracy of  the train_dataset
Accuracy_y_train = accuracy_score(y_train, y_train_pred)
```

```python
print(Accuracy_y_train)
```

```python
Conf_Mat_train = confusion_matrix(y_train, y_train_pred)
```

```python
Conf_Mat_train
```

```python
# Making Predictons on the test dataset
y_test_pred = model.predict(x_test)
```

```python
Accuracy_y_test = accuracy_score(y_test, y_test_pred)
```

```python
print(Accuracy_y_test)
```
