## Cleaning Transformations to apply to the dataset

```python
""" Deleting redundant columns.
Renaming the columns.
Dropping duplicates.
Remove the NaN values from the dataset.
Cleaning individual columns.
Checking for some more transformations."""
```

### Deleting redundant columns:

```python
# Loading the columns
Airbnb.columns
```

```python
# We keep all columns untill last review and drop the rest
cols_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license', 'id']
```

```python
cols_to_keep  = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
```

```python
# Checking len of both columns: 7 columns
len(cols_to_drop)
```

```python
# columns to keep = 19
len(cols_to_keep)
```

```python
# dropping columns:
Airbnb.drop(columns = cols_to_drop, axis = 1, inplace = True)
```

```python
Airbnb.head() # lodding 5 rows of remaining data
```

```python
# New shape:
Airbnb.shape
```

### Renaming the columns:

```python
Airbnb.columns
```

```python
Airbnb.rename({"NAME":"name"}, axis = 1, inplace = True)
```

```python
Airbnb.head()
```

### Dropping duplicates.

```python
# Checking for duplicates:
Airbnb.loc[Airbnb.duplicated(keep = False)]
```

```python
# Checking for the number of duplicates and non-duplicates
Airbnb.duplicated(keep = 'first').value_counts()
```

```python
# Duplicates: 541
# Non-duplicates: 102058
```

```python
# Dropping off duplicates
Airbnb.drop_duplicates(keep = 'first', inplace = True)
```

```python
Airbnb.loc[Airbnb.duplicated()] # Checks for duplicates
```

### All duplicates are removed finally

### Removing NaN Values:

```python
# Checking for all NaN values:
Airbnb.isna().sum().sort_values()
```

```python
# Checking for Actual size of each feature column before performing NaN removal
Airbnb.info()
```

```python
Airbnb.name.shape
```

```python
Airbnb.name.isna().sum()
```

```python
## First dropping off the last review
Airbnb.drop(columns = ["last review"], inplace = True, axis = 1)
```

```python
Airbnb.isna().sum().sort_values()
```

```python
## Dropping off some NaN values:
Airbnb.dropna(how ='any', inplace = True, axis = 0)
```

```python
Airbnb.isna().sum()
```

### Cleaning individual columns.

```python
# Checking the columns
Airbnb.columns
```

```python
Airbnb.head()
```

```python
# 1. Making the column: "host_identity_verified"
Airbnb.host_identity_verified
```

```python
# Changing all to upper_case 
Airbnb.host_identity_verified.str.upper()
```

```python
Airbnb["host_identity_verified"] = Airbnb["host_identity_verified"].str.upper()
```

```python
Airbnb.head() # Column changed to uppercase
```

### Renaming feature: instant_bookable to boolean

```python
Airbnb["instant_bookable"].value_counts()
```

```python
# Changing from object - boolean
Airbnb["instant_bookable"] = Airbnb["instant_bookable"].astype("boolean")
```

```python
Airbnb.instant_bookable.dtype
```

### Changing all False as 0 and True as 1

```python
Airbnb["instant_bookable"] = Airbnb["instant_bookable"].apply(lambda x: 1 if x == True else 0)
```

```python
Airbnb.head()
```

### Checking for number of number of instant_bookable and not instant_bookable

```python
Airbnb["instant_bookable"].value_counts()
```

### This implies we have: No istant_bookable of 49878 and instant_bookable as 49464 

```python
Airbnb.head()
```

```python
# Resetting the index of the dataset:
Airbnb.reset_index(inplace = True)
```

```python
Airbnb.head()
```

```python
# Checking the datatype of the price feature:
Airbnb["price"].dtype
```

```python
Airbnb["price"] = Airbnb["price"].str.replace("$", "")
```

```python
Airbnb["price"] = Airbnb["price"].str.replace(",", "")
```

```python
Airbnb["price"] = Airbnb["price"].str.strip()
```

```python
# Changing to int:
Airbnb["price"]
```

```python
# Applying same thing for service fees
```

```python
Airbnb["service fee"] = Airbnb["service fee"].str.replace("$", "")
```

```python
Airbnb["service fee"] = Airbnb["service fee"].str.replace(",", "")
```

```python
Airbnb["service fee"] = Airbnb["service fee"].str.strip()
```

```python
Airbnb["service fee"]
```

```python
Airbnb.head()
```

```python
Airbnb["price"].dtype
```

```python
Airbnb["service fee"].dtype
```

```python
Airbnb.info()
```

```python
Airbnb = Airbnb.astype({"price": "int", "service fee": "int"})
```

```python
Airbnb.dtypes
```
